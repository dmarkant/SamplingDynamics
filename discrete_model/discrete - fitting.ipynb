## Fitting a single problem

```python
import pandas as pd
```

```python
data = pd.read_csv('../dfe_by_game.csv')
```

```python
data[data['gamble_lab']=='S10G1'].head(10)
```

```python
gdata = data[data['gamble_lab']=='S12G1']
```

```python
tofit = []
for pair in gdata[['decision', 'samplesize']].values:
    choice = 0 if pair[0]=='L' else 1
    samplesize = pair[1]
    tofit.append([choice, samplesize])
tofit = np.array(tofit)
```

```python
ss_A = tofit[np.where(tofit[:,0]==0)[0],1]
ss_B = tofit[np.where(tofit[:,0]==1)[0],1]
max_t = np.max([np.max(ss_A), np.max(ss_B)]) + 1
```

```python
fig, ax = plt.subplots()
ax.hist(tofit[np.where(tofit[:,0]==0)[0],1], bins=50)
ax.hist(tofit[np.where(tofit[:,0]==1)[0],1], bins=50)
plt.show()
```

```python
# fitting theta and delta
pars = {'data': tofit,
        'z_width': .5,
        'max_T': max_t,
        'fitting': ['theta', 'delta'],
        'verbose': False}

loglik([5., 0.], pars)
```

```python
f = minimize(loglik, [5., 0.], (pars,), method='Nelder-Mead')
```

```python
f
```

```python
bf_pred = run({'theta': f['x'][0],
               'delta': f['x'][1],
               'z_width': .5,
               'max_T': max_t,
               'verbose': True})
```

```python
T = bf_pred['T']
fig, ax = plt.subplots(1, 2, figsize=(15,5))
ax[0].plot(T, bf_pred['resp_prob_t'][:,0], label='A')
ax[0].set_title('Chose A')
axhA = ax[0].twinx()
axhA.hist(ss_A, bins=20, alpha=.3)

ax[1].plot(T, bf_pred['resp_prob_t'][:,1], label='B')
ax[1].set_title('Chose B')
axhB = ax[1].twinx()
axhB.hist(ss_B, bins=20, alpha=.3)
plt.show()
```

```python
# fitting theta and delta
pars = {'data': tofit,
        'max_T': max_t,
        'fitting': ['theta', 'delta', 'z_width'],
        'verbose': False}

loglik([5., 0., .5], pars)
```

```python
f = minimize(loglik, [5., 0., 0.5], (pars,), method='Nelder-Mead')
```

## All problems, single drift

```python
gamble_lab = data['gamble_lab'].unique()
```

```python
gid = gamble_lab[0]
```

```python
fit_pars = {}
fit_results = {}
for gid in gamble_lab:
    gdata = data[data['gamble_lab']==gid]
    tofit = []
    for pair in gdata[['decision', 'samplesize']].values:
        choice = 0 if pair[0]=='L' else 1
        samplesize = pair[1]
        tofit.append([choice, samplesize])
    tofit = np.array(tofit)
    
    ss_A = tofit[np.where(tofit[:,0]==0)[0],1]
    ss_B = tofit[np.where(tofit[:,0]==1)[0],1]
    max_t = np.max([np.max(ss_A), np.max(ss_B)]) + 1
    
    # fitting theta and delta
    pars = {'data': tofit,
            'max_T': max_t,
            'fitting': ['theta', 'delta', 'z_width'],
            'verbose': False}
    
    #print loglik([5., 0., .5], pars)
    fit_pars[gid] = pars
    fit_results[gid] = minimize(loglik, [5., 0., 0.5], (pars,), method='Nelder-Mead')
    print gid
    print fit_results[gid]
```

```python
for gid in gamble_lab:
    pars = fit_pars[gid]
    
    for i, p in enumerate(pars['fitting']):
        pars.update({p: fit_results[gid]['x'][i]})

    bf_pred = run(pars)

    tofit = pars['data']
    ss_A = tofit[np.where(tofit[:,0]==0)[0],1]
    ss_B = tofit[np.where(tofit[:,0]==1)[0],1]
    
    fig, ax = plt.subplots(1, 2, figsize=(10,3))
    ax[0].plot(bf_pred['T'], bf_pred['resp_prob_t'][:,0], label='A')
    ax[0].set_title('%s: Chose A' % gid)
    axhA = ax[0].twinx()
    axhA.hist(ss_A, bins=bf_pred['T'], alpha=.9)
    
    ax[1].plot(bf_pred['T'], bf_pred['resp_prob_t'][:,1], label='B')
    ax[1].set_title('%s: Chose B' % gid)
    axhB = ax[1].twinx()
    axhB.hist(ss_B, bins=bf_pred['T'], alpha=.9)
    plt.show()    
```

```python
fig, ax = plt.subplots(figsize=(15,4))
bf_theta = [fit_results[gid]['x'][0] for gid in gamble_lab]
ax.bar(range(len(bf_theta)), bf_theta)
plt.show()
```
