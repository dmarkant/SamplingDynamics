```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mypy.explib.frey2014 import frey2014
from model_natmean import model_natmean
%load_ext autoreload
%autoreload 2
```

```python
subjdata = frey2014.sampledata_by_subject()
```

```python
sid = 115
```

```python
sdata = subjdata[sid]
#sdata = filter(lambda x: np.max(x['outcomes']) > 0, subjdata[sid])
```

```python
ret = model_natmean.fit_subject_across_gambles(sdata, 
                                               {}, 
                                               ['theta', 'eval_pow', 'eval_crit', 'z_sd', 'rho'])
```

```python
ret
```

```python
states = []
for sd in sdata:
    pars = {'data': sd}
    for p in ret['bf_par']:
        pars[p] = ret['bf_par'][p]
    result = model_natmean.run(pars)
    states.append([sd['choice'], result['pref']])
```

```python
states_A = []
states_B = []
for i, gstate in enumerate(states):
    if subjdata[sid][i]['choice']==0:
        states_A.append(gstate)
    else:
        states_B.append(gstate)
```

```python
states_A = [st[1] for st in filter(lambda s: s[0]==0, states)]
states_B = [st[1] for st in filter(lambda s: s[0]==1, states)]
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20,5))
for i, gstate in enumerate(states_A):
    ax[0].plot(gstate, color='blue', alpha=.2)
ax[0].plot([0, 30], [ret['bf_par']['theta'], ret['bf_par']['theta']], color='black')
ax[0].plot([0, 30], [-ret['bf_par']['theta'], -ret['bf_par']['theta']], color='black')
for i, gstate in enumerate(states_B):
    ax[1].plot(gstate, color='red', alpha=.2)
ax[1].plot([0, 30], [ret['bf_par']['theta'], ret['bf_par']['theta']], color='black')
ax[1].plot([0, 30], [-ret['bf_par']['theta'], -ret['bf_par']['theta']], color='black')
plt.show()
```

For all subjects, plot the natural mean over time for each gamble, grouped by which option was chosen. This will give a pretty good idea of those subjects that are likely to be fit well with this model
