```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from mypy.explib.hau2008 import hau2008
import model_natmean as seqnm
%load_ext autoreload
%autoreload 2
```

```python
df_samples, df_choices = hau2008.load_study(1)
```

```python
SUBJ = df_samples['subject'].unique()
```

```python
results = {}
for pid in range(6):
    results[pid] = []
    for sid in df_choices[df_choices['problem']==(pid+1)]['subject'].unique():

        sampledata, choice = hau2008.get_subj_data(1, sid, pid)

        pars = {'data': {'sampledata': sampledata,
                         'choice': choice},
                'eval_pow': 1.,
                'theta': 20}
        r = seqnm.run(pars)
        r['choice'] = choice
        r['sampledata'] = sampledata
        results[pid].append(r)
```

```python
fig, ax = plt.subplots(6, 1, figsize=(10,20))
for pid in range(6):
    choseA = filter(lambda r: r['choice']==0, results[pid])
    choseB = filter(lambda r: r['choice']==1, results[pid])
    for r in choseA:
        ax[pid].plot(r['pref'], color='red', alpha=.7)
    for r in choseB:
        ax[pid].plot(r['pref'], color='blue', alpha=.7)
```

```python
results = {}
for pid in range(6):
    results[pid] = []
    for sid in df_choices[df_choices['problem']==(pid+1)]['subject'].unique():

        sampledata, choice = hau2008.get_subj_data(1, sid, pid)

        pars = {'data': {'sampledata': sampledata,
                         'choice': choice},
                'eval_pow': .75,
                'theta': 20}
        r = seqnm.run(pars)
        r['choice'] = choice
        r['sampledata'] = sampledata
        results[pid].append(r)
```

```python
# combine all data together
fitdata = []
for pid in range(6):
    for sid in df_choices[df_choices['problem']==(pid+1)]['subject'].unique():
        sampledata, choice = hau2008.get_subj_data(1, sid, pid)
        fitdata.append({'pid': pid,
                        'sid': sid,
                        'sampledata': sampledata,
                        'choice': choice})
```

```python
pars = {'data': fitdata,
        'fitting': ['theta'],
        'eval_pow': 1.,
        'theta': 20}
for th in [10, 50., 100., 200., 1000.]:
    print th, seqnm.loglik_across_gambles([th], pars)
```
