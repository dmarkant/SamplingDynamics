### Natmean with fixed switching probability

```python
def run_all_subj_for_game(study, gid, model, pars):
    gdata = {sid: hau2008.get_subj_data(study, sid, gid) for sid in SUBJ}

    # use only those participants who completed this problem
    sset = filter(lambda sid: gdata[sid]['choice']!=None, SUBJ)
    
    results = {}
    for sid in sset:
        pars['data'] = gdata[sid]
        results[sid] = model.run(pars)
    return results
```

```python
study = 1
results = {}
for gid in range(6):
    results[gid] = run_all_subj_for_game(study, gid, model_nmp, {'rho': .5,
                                                                 'eval_pow': .1,
                                                                 'mu': 10,
                                                                 'sd': 2})
```

```python
fig, ax = plt.subplots(6, 1, figsize=(15,15))
for gid in range(6):

    for sid in results[gid]:
        gdata = hau2008.get_subj_data(study, sid, gid)
        ax[gid].plot(results[gid][sid]['pref'], color=('blue' if gdata['choice']==0 else 'red'))
```

```python
def plot_result(sid):
    if sid in fitresults_nm:

        fr = fitresults_nm[sid]
        states = []
        for sd in subjdata[sid]:
            pars = {'data': sd}
            for p in fr['bf_par']:
                pars[p] = fr['bf_par'][p]
            result = model_natmean.run(pars)
            states.append([sd['choice'], result['pref']])
        states_A = [st[1] for st in filter(lambda s: s[0]==0, states)]
        states_B = [st[1] for st in filter(lambda s: s[0]==1, states)]
        max_t = max([sd['sampledata'].size for sd in subjdata[sid]])
        
        fig, ax = plt.subplots(1, 2, figsize=(20,5))
        for i, gstate in enumerate(states_A):
            ax[0].plot(gstate, color='blue', alpha=1)
        ax[0].plot([0, max_t+1], [ fr['bf_par']['theta'],  fr['bf_par']['theta']], '-.', color='blue')
        ax[0].plot([0, max_t+1], [-fr['bf_par']['theta'], -fr['bf_par']['theta']], '--', color='red')
        ax[0].set_title('Subject %s, A choices' % sid)
        
        for i, gstate in enumerate(states_B):
            ax[1].plot(gstate, color='red', alpha=1)
        ax[1].plot([0, max_t+1], [ fr['bf_par']['theta'],  fr['bf_par']['theta']], '-.', color='blue')
        ax[1].plot([0, max_t+1], [-fr['bf_par']['theta'], -fr['bf_par']['theta']], '--', color='red')
        ax[1].set_title('Subject %s, B choices' % sid)
        plt.show()
```
