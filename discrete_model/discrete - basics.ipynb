```python
import numpy as np
from scipy import linalg
from numpy.linalg import matrix_power
```

```python
def transition_probs(delta, sigma, tau):
    p_down = .5 * (1 - (delta/sigma) * np.sqrt(tau))
    p_up = .5 * (1 + (delta/sigma) * np.sqrt(tau))
    p_stay = 1 - p_down - p_up
    return [p_down, p_stay, p_up]
```

```python
# Diederich (1997), p.263
# Note that ((delta - sdw)/sigma) must be within +- 1/sqrt(tau) 
# to produce a probability matrix.. should build in this constraint

# Diederich (2003) form might have this already..

def transition_probs_state_dependent(v, delta, sigma, tau, gamma):
    # state-dependent weight depends on current state and gamma
    sdw = gamma * v
    p_down = .5 * (1 - ((delta - sdw)/sigma) * np.sqrt(tau))
    p_up = .5 * (1 + ((delta - sdw)/sigma) * np.sqrt(tau))
    p_stay = 1 - p_down - p_up
    return [p_down, p_stay, p_up]
```

```python
# free parameters
theta = 5   # criterion
dv = .25    # step size between states
delta = .2  # drift parameter
sigma = 1.  # diffusion parameter
gamma = 0.  # state-dependent weighting of drift

tau = (dv**2)/(sigma**2) # time step

# state space
V = np.arange(-theta, theta+dv, dv)
vi = range(len(V))
m = len(V)

# starting position, not including absorbing states!
zhw = (m - 2)/3
Z = np.zeros(m - 2)
Z[(len(Z)/2-zhw):(len(Z)/2+zhw+1)] = 1.
Z = np.matrix(Z/float(sum(Z)))

# transition probabilities based on drift
# and diffusion parameters
tr_p = transition_probs(delta, sigma, tau)

# construct the transition probability matrix
tm = np.zeros((m, m), float)
tm[0,0] = 1.
tm[m-1,m-1] = 1.
for i in range(1, m - 1):
    #tm[i, (i-1):(i+2)] = tr_p
    tm[i, (i-1):(i+2)] = transition_probs_state_dependent(i*dv, delta, sigma, tau, gamma)

# construct matrix in PQR form
tm_pqr = np.zeros((m, m), float)
tm_pqr[0,0] = 1.
tm_pqr[1,1] = 1.
vi_pqr = []
start = np.array([[0, m - 1], range(1, m - 1)])
for outer in start:
    for inner in outer:
        vi_pqr.append(inner)
vi_pqr = np.array(vi_pqr)
V_pqr = V[vi_pqr]
for i in range(1, m - 1):
    row = np.where(V_pqr==V[i])[0][0]
    ind_pqr = np.array([np.where(V_pqr==V[i-1])[0][0], np.where(V_pqr==V[i])[0][0], np.where(V_pqr==V[i+1])[0][0]])
    #tm_pqr[row, ind_pqr] = tr_p
    tm_pqr[row, ind_pqr] = transition_probs_state_dependent(i*dv, delta, sigma, tau, gamma)

Q = tm_pqr[2:,2:]
R = np.matrix(tm_pqr[2:,:2])
I = np.eye(m - 2)
```

```python
IQ = np.matrix(linalg.inv(I - Q))
resp_prob = Z * (IQ * R)
dt = (Z * (IQ * IQ) * R) / resp_prob

print 'response prob:', resp_prob
print 'expected time | chosen alternative:', dt
```

```python
# when do you want to evaluate predictions?
dt = 1.
T = np.arange(0, 100, dt)
```

```python
# number of steps completed at each time point
N = map(int, np.floor(T/tau) + 1)
```

```python
# how does response prob change over time?
#N = range(1, 1000)
resp_prob_t = np.array([Z * (matrix_power(Q, n - 1) * R) for n in N]).reshape((len(N), 2))

# this is the problem.. has to be based on all steps
resp_prob_N = np.array([Z * (matrix_power(Q, n - 1) * R) for n in range(1, N[-1])]).reshape((N[-1]-1, 2))
resp_prob_cump = resp_prob_N.cumsum(axis=0)
```
