```python
fitresults_nm = {}
cols = ['theta', 'eval_pow', 'eval_crit', 'z_sd', 'rho', 'nllh_sampling', 'nllh_choice', 'bic', 'success']
fitresults_nm_table = pd.DataFrame(index=SUBJ, columns=cols)
```

```python
for sid in SUBJ[20:]:
    print sid
    fitresults_nm[sid] = model_natmean.fit_subject_across_gambles(subjdata[sid], 
                                                   {}, 
                                                   ['theta', 'eval_pow', 'eval_crit', 'z_sd', 'rho'])
    
    pars = {'data': subjdata[sid],
            'fitting': []}
    for p in fitresults_nm[sid]['bf_par']:
        pars[p] = fitresults_nm[sid]['bf_par'][p]
    fitresult = model_natmean.nloglik_across_gambles([], pars)

    fitresults_nm_table.loc[sid] = [fitresults_nm[sid]['bf_par']['theta'], 
                                    fitresults_nm[sid]['bf_par']['eval_pow'],
                                    fitresults_nm[sid]['bf_par']['eval_crit'],
                                    fitresults_nm[sid]['bf_par']['z_sd'],
                                    fitresults_nm[sid]['bf_par']['rho'],
                                    fitresult['llh_sampling'], fitresult['llh_choice'], 
                                    fitresults_nm[sid]['bic'], fitresults_nm[sid]['success']]
```

```python
fitresults_nm_table.to_csv('hau_natmean.csv')
fitresults_nm_table
```

```python
sid = 8
fr = fitresults_nm[sid]
states = []
for sd in subjdata[sid]:
    pars = {'data': sd}
    for p in fr['bf_par']:
        pars[p] = fr['bf_par'][p]
    result = model_natmean.run(pars)
    states.append([sd['choice'], result['pref']])
states_A = [st[1] for st in filter(lambda s: s[0]==0, states)]
states_B = [st[1] for st in filter(lambda s: s[0]==1, states)]
```

```python
max_t = max([sd['sampledata'].size for sd in subjdata[sid]])
fig, ax = plt.subplots(1, 2, figsize=(20,5))
for i, gstate in enumerate(states_A):
    ax[0].plot(gstate, color='blue', alpha=1)
ax[0].plot([0, max_t+1], [ fr['bf_par']['theta'],  fr['bf_par']['theta']], '-', color='blue')
ax[0].plot([0, max_t+1], [-fr['bf_par']['theta'], -fr['bf_par']['theta']], '--', color='red')
for i, gstate in enumerate(states_B):
    ax[1].plot(gstate, color='red', alpha=1)
ax[1].plot([0, max_t+1], [ fr['bf_par']['theta'],  fr['bf_par']['theta']], '-', color='blue')
ax[1].plot([0, max_t+1], [-fr['bf_par']['theta'], -fr['bf_par']['theta']], '--', color='red')
plt.show()
```

```python
def plot_result(sid):
    if sid in fitresults_nm:

        fr = fitresults_nm[sid]
        states = []
        for sd in subjdata[sid]:
            pars = {'data': sd}
            for p in fr['bf_par']:
                pars[p] = fr['bf_par'][p]
            result = model_natmean.run(pars)
            states.append([sd['choice'], result['pref']])
        states_A = [st[1] for st in filter(lambda s: s[0]==0, states)]
        states_B = [st[1] for st in filter(lambda s: s[0]==1, states)]
        max_t = max([sd['sampledata'].size for sd in subjdata[sid]])
        
        fig, ax = plt.subplots(1, 2, figsize=(20,5))
        for i, gstate in enumerate(states_A):
            ax[0].plot(gstate, color='blue', alpha=1)
        ax[0].plot([0, max_t+1], [ fr['bf_par']['theta'],  fr['bf_par']['theta']], '-.', color='blue')
        ax[0].plot([0, max_t+1], [-fr['bf_par']['theta'], -fr['bf_par']['theta']], '--', color='red')
        ax[0].set_title('Subject %s, A choices' % sid)
        
        for i, gstate in enumerate(states_B):
            ax[1].plot(gstate, color='red', alpha=1)
        ax[1].plot([0, max_t+1], [ fr['bf_par']['theta'],  fr['bf_par']['theta']], '-.', color='blue')
        ax[1].plot([0, max_t+1], [-fr['bf_par']['theta'], -fr['bf_par']['theta']], '--', color='red')
        ax[1].set_title('Subject %s, B choices' % sid)
        plt.show()
```
