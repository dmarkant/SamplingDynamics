```python
def samplesize(study, gid):
    df_samples, df_choices = load_study(study)
    return df_samples[df_samples['problem']==(gid+1)].sort('subject').groupby('subject').apply(lambda x: len(x)).values
```

```python
df_samples, df_choices = load_study1()
```

```python
def choices(study, gid):
    df_samples, df_choices = load_study(study)
    c = df_choices[df_choices['problem']==(gid+1)].sort('subject')['choice'].values
    if study==1 and gid==0:
        return c
    else:
        return np.abs(c - 1)
```

```python
df_samples, df_choices = load_study(2)
```

# Study 1

```python
ss = [samplesize(1, i) for i in range(6)]

fig, ax = plt.subplots(1, 6, figsize=(20,4))
for gid in range(6):
    ax[gid].hist(ss[gid], bins=20)
    ax[gid].set_ylim(0, 15)
plt.show()
```

```python
fitdata = []
for gid in range(6):
    tofit = np.transpose((np.array(choices(1, gid)), np.array(samplesize(1, gid)), np.ones(len(choices(1, gid)))))
    
    ss_A = tofit[np.where(tofit[:,0]==0)[0],1]
    ss_B = tofit[np.where(tofit[:,0]==1)[0],1]
    max_t = np.max([np.max(ss_A), np.max(ss_B)]) + 1
    
    fitdata.append({'gid': gid,
                    'options': get_options(1, gid),
                    'max_t': max_t,
                    'samplesize': tofit})
```

### delta, theta

```python
pars = {'data': fitdata,
        'fitting': ['delta', 'theta'],
        'bounds': [[-np.inf, np.inf],
                   [1, np.inf]]
}
init = [np.random.random(), 5.]
f = minimize(loglik_across_gambles, init, (pars,), method='Nelder-Mead')
```

```python
print f
print 'bic: ', bic(f, pars)
```

```python
p_resp = []
p_stop = []
prop_choices_predicted = []
for i, fd in enumerate(fitdata):
    pred_pars = {'data': fd['samplesize'],
                 'max_T': fd['max_t'],
                 'options': fd['options']}
    for par_i, par_name in enumerate(pars['fitting']):
        pred_pars[par_name] = f['x'][par_i]
    result = run(pred_pars)
    p_resp.append(result['resp_prob'][1])
    
    exp_tA = np.sum([t * pt for (t, pt) in enumerate(result['p_stop_t'][:,0])]) * result['resp_prob'][0]
    exp_tB = np.sum([t * pt for (t, pt) in enumerate(result['p_stop_t'][:,1])]) * result['resp_prob'][1]
    p_stop.append(exp_tA + exp_tB)
    
    matched = [choice==np.argmax(result['resp_prob_t'][t]) for (choice, t, grp) in pred_pars['data']]
    prop_choices_predicted.append(np.sum(1*matched)/float(len(matched)))

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(range(1,7), [np.mean(choices(1, pid)) for pid in range(6)], '-o', label='observed')
ax.plot(range(1,7), p_resp, '-x', label='best fit model')
ax.set_xlim(0, 7)
ax.set_ylim(0, 1)
ax.legend()
plt.show()
print np.mean(prop_choices_predicted)
```
