# Sampling dynamics

Summary of data:

- Two groups of participants: (1) 35 younger adults, (2) 35 older adults
- Each participant completed 3 sessions a day for 7 consecutive days
- Each session included 4 decision problems, 2 in gain domain and 2 in loss domain
- At end of each session, received feedback about total gains/losses

Four sets of options:

1. low expected value, low variance (LL)
2. low expected value, high variance (LH)
3. high expected value, low variance (HL)
4. high expected value, high variance (HH)

Decision problems involved pairing

1. HH vs. LL
2. HL vs. LH


## Questions

Main question is whether sequential sampling models can predict sampling dynamics, including final choice, overall sample size and decisions to switch between options (leaving aside questions about RTs at the moment, since those are not available for this dataset). This kind of modeling approach would be useful under the assumption that people make ongoing decisions over the course of sampling that are influenced by the outcomes of previous decisions. 

First goal is to evaluate whether certain aspects of sampling/choice dynamics can't be explained by fixed strategies or problem features, which may suggest a decision process that can be modeled with sequential sampling framework. Aim to answer the following questions:

1. [Switching at choice](#Switching-at-choice) How often do people choose a different option from the most recently sampled one? (As discussed on 18.07.2014, common switches *at choice*--picking the option that was not sampled most recently--suggests some kind of dependency in accumulation)

1. [Switching during sampling](#switching) How often do people switch between options during sampling? How does switching frequency relate to overall sample size? 

1. [How common are extreme strategies like switching between options on every trial?](#Detecting-switching-strategies)

1. How much variability exists in the timing of switch decisions (assuming relatively low switch rate)?

1. Can switches be predicted by simple rules, like 50/50 allocation, or planned sampling runs (e.g., sample N times, then switch)?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def stay_or_switch(gamedf):
    r = [-1]
    for i in range(1, len(gamedf.index)):    
        # check if option is same or different from previous trial
        prev = gamedf.ix[gamedf.index[i-1]]['sample_opt']
        curr = gamedf.ix[gamedf.index[i]]['sample_opt']
        if prev == curr:
            r.append(0)
        else:
            r.append(1)
    
    return np.array(r)

def stop_trial(gamedf):
    r = np.zeros(len(gamedf), int)
    r[-1] = 1
    return r

def num_switches(gamedf):
    return len(gamedf[gamedf['switch']==1])

def switching_frequency(gamedf):
    stay = len(gamedf[gamedf['switch']==0])
    switch = len(gamedf[gamedf['switch']==1])
    if (stay + switch) > 0:
        return switch / float(stay + switch)
    else:
        return 0

def sample_size(gamedf):
    return len(gamedf)

filename = 'dfe.csv'
df = pd.read_csv(filename)
df = df.sort(['partid', 'gamble_lab', 'sample_ind'])
SUBJ = df['partid'].unique()
```

```python
result = df[['partid', 'gamble_lab', 'sample_opt']].groupby(['partid', 'gamble_lab']).apply(lambda x: stay_or_switch(x))
df['switch'] = list(flatten(result.values))
```

```python
switchfreq  = df.groupby(['partid', 'gamble_lab']).apply(switching_frequency).values
switchcount = df.groupby(['partid', 'gamble_lab']).apply(num_switches).values
```

```python
gdf = df.groupby(['partid', 'gamble_lab']).size().reset_index()
gdf.columns = ['partid', 'gamble_lab', 'samplesize']
gdf['switchfreq'] = switchfreq
gdf['switchcount'] = switchcount
```

```python
gambles = df[df['partid']==1]['gamble_lab'].unique()
GAMBLES = {}
for g in gambles:
    H = sort(df[(df['gamble_lab']==g) & (df['sample_opt']=='H')]['sample_out'].unique())
    L = sort(df[(df['gamble_lab']==g) & (df['sample_opt']=='L')]['sample_out'].unique())
    GAMBLES[g] = {'H': H, 'L': L}
```

### Switching at choice

How often do people choose an option that is not the most recently sampled option?

```python
def leadup_count(gamedf):
    choice = gamedf['decision'].values[0]
    samples = gamedf['sample_opt']
    count_same = 0
    count_other = 0
    for s in samples:
        if s == choice:
            count_same += 1
            count_other = 0
        else:
            count_same = 0
            count_other += 1
    return np.array([count_same, count_other])

counts = {}
for sid in SUBJ:
    
    gambles = df[df['partid']==sid]['gamble_lab'].unique()
    counts[sid] = []
    for gid in gambles:
        g = df[(df['partid']==sid) & (df['gamble_lab']==gid)]
        counts[sid].append(leadup_count(g))
    counts[sid] = np.array(counts[sid])
    
bicounts = {}
for sid in SUBJ:
    bicounts[sid] = np.array([p[0] - p[1] for p in counts[sid]])
```

The x-axis is the number of samples in the final sampling streak (negative values indicate that the samples were drawn from the unchosen option). The height of each bar indicates the number of problems in which that streak size occurred. 

Red bars indicate when the most recently sampled option was chosen (no switch at choice), blue bars indicate when the most recently sampled option was NOT chosen (switch at choice). Adding up all of the blue bars gives an indication of how often a participant chose an option that wasn't sampled most recently. 

```python
jit = .1
cols = 5
rows = len(SUBJ)/cols 
if len(SUBJ)%cols > 0: rows += 1

fig, ax = plt.subplots(rows, cols, figsize=(16,40))
for i in range(len(SUBJ)):
    axi = ax[i/cols][i%cols]
    sid = SUBJ[i]    
    axi.hist([c for c in bicounts[sid] if c > 0], bins=range(max(bicounts[sid])+1), alpha=0.5, color='red', histtype='stepfilled')
    axi.hist([c for c in bicounts[sid] if c < 0], bins=range(min(bicounts[sid])-1, 0), alpha=0.5, color='blue', histtype='stepfilled')
    axi.set_title('subj %s' % sid)
    axi.set_xlim(min([min(bicounts[sid]), -10]), max([max(bicounts[sid]), 10]))
plt.tight_layout()
plt.show()
```

Some participants (e.g., 130) switch on nearly every trial during sampling, and as a result their final streaks only include a single sample (either from the chosen or unchosen option). But note that among other participants there is greater variability in the final streak length -- including when the final streak was drawn from the unchosen option. This means that it was quite common for people to sample from an option 5-10 times only to choose the other option.

```python
def switch_at_choice_after_worst(sid):
    gambles = df[df['partid']==sid]['gamble_lab'].unique()
    result = []
    for i, gid in enumerate(gambles):
        if gid in GAMBLES:
            #print gid
            switch_at_choice = 1 if bicounts[sid][i] < 0 else 0
            gamedf = df[(df['partid']==sid) & (df['gamble_lab']==gid)]
            
            choice = gamedf['decision'].values[0]
            last_sample = gamedf['sample_opt'].values[-1]
            if last_sample in ['L', 'H']:
            
                # for the option that was sampled last, get all observed outcomes
                #outcomes = sort(gamedf[gamedf['sample_opt']==last_sample]['sample_out'].unique())
                #print outcomes
                
                last_outcome = gamedf['sample_out'].values[-1]
                
                #print i
                #print last_sample, last_outcome
                #print GAMBLES[gid][last_sample]
                #print '_'*10
                
                rank_last_outcome = np.argwhere(GAMBLES[gid][last_sample]==last_outcome)[0,0]
                if rank_last_outcome == 0:
                    worst = 1
                else:
                    worst = 0
                result.append([worst, switch_at_choice])
    return np.array(result)
```

```python
# conditional probs
# - first column: frequency of switching when the most recent outcome was not the worst
# - second column: frequency of switching when the most recent outcome was the worst
cp_switch_after_worst = []
for sid in SUBJ:
    arr = switch_at_choice_after_worst(sid)
    cp_switch_after_worst.append([np.mean(arr[arr[:,0]==0][:,1]), np.mean(arr[arr[:,0]==1][:,1])])
cp_switch_after_worst = np.array(cp_switch_after_worst)    
```

### Detecting switching strategies

Averaging the switching rate across problems is problematic since the maximum rate depends on the number of samples drawn. A better way to detect alternation is to look at how the number of switches depends on the number of samples in a game. In the plots below, each cross is a different problem, and the diagonal line indicates the predicted pattern from switching on every trial.

```python
jit = .1
cols = 5
rows = len(SUBJ)/cols 
if len(SUBJ)%cols > 0: rows += 1

fig, ax = plt.subplots(rows, cols, figsize=(16,40))
for i in range(len(SUBJ)):
    axi = ax[i/cols][i%cols]
    s = SUBJ[i]
    x = gdf[gdf['partid']==s][['samplesize', 'switchcount']]
    ss = [val + random.uniform(-jit, jit) for val in x['samplesize']]
    sc = [val + random.uniform(-jit, jit) for val in x['switchcount']]
    axi.plot(ss, sc, 'b+')
    
    axi.plot([1, max(ss)], [0, (max(ss)-1)], 'k--')
    axi.plot([0, max(ss)], [0, 0], 'k--')
    axi.set_xlim(0, np.max(ss))
    axi.set_ylim(-5, np.max(ss))
    axi.set_title('subj %s' % s)
plt.tight_layout()
plt.show()
```

People who fall on the diagonal line have a strategy of alternating on every sample, and so it's unlikely that anything about their switching behavior is dependent on the outcomes they observe or features of the gamble. 

People who fall on a horizontal line have a relatively low switching frequency across all problems; for those cases it may still be possible to predict switch points based on their sampling experience.

Note that there are relatively few participants who fall midway between these two extremes (this is also seen in the bimodal distribution of average switching frequency in the plot above). If people used fixed-interval switching strategies (like 'switch every 3rd sample') then that should appear as a diagonal line with shallower slope (e.g., some hint of this in subject 8) -- but this doesn't seem to be a common pattern.

```python
def plot_switch_sequences(sid, ax):
    gambles = df[df['partid']==sid]['gamble_lab'].unique()
    switch_sequences = []
    for gid in gambles:
        g = df[(df['partid']==sid) & (df['gamble_lab']==gid)]
        switch_sequences.append(g['switch'].values)
    switch_sequences.sort(key=lambda x: len(x))
    
    for i, seq in enumerate(switch_sequences):
        n = len(seq)
        x = np.arange(n) - (n)/2.
        if n > 1:
            ax.plot(x[1:], (i+seq)[1:], 'k-')
        ax.plot([0, 0], [0, i+2], 'k--')
    ax.set_ylim(-5, i+5)
    ax.set_title('subj %s' % sid)
    ax.set_xlabel('sample position\nrelative to middle of sequence')
    ax.set_ylabel('problem index (sorted by # samples)')
    return ax
```

Aside from alternation, the other extreme (fixed) strategy would be to always use the same allocation of samples to options (e.g., split 50-50), in which case switches should occur in at consistent points in the sequence. The following plot shows a few participants with low switching frequency and low variability in switch points (usually near the midpoint of a sequence). Each horizontal line is a different problem, with switches indicated by peaks in the line. Subject 1 is an extreme example (almost always taking the same # of samples and switching at exactly the midpoint), whereas subject 3 is somewhat more variable in when they decide to switch (though switches are still clustered around the midpoint).

```python
fig, ax = plt.subplots(1, 3, figsize=(16,8))
ax[0] = plot_switch_sequences(1, ax[0])
ax[1] = plot_switch_sequences(3, ax[1])
ax[2] = plot_switch_sequences(5, ax[2])
plt.show()
```

These plots show switches for participants where a fixed strategy is less apparent (although as noted above, subj 8 seems to explore a few fixed-interval switching strategies).
