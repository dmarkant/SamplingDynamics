# Hau et al (2008)

In this paper there were three experiments that used the original set of problems from Hertwig et al (2004). 

- Exp 1: Direct replication of Hertwig et al. (2004)
- Exp 2: Increased stakes 
- Exp 3: Increased sample size (fixed at 100 per problem)

Here going to fit the model to the results of Experiment 1 and 2. First will fit models separately, then will test for a difference between Experiment 1 and 2 by combining those datasets.

### Proportion choosing H

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(range(1,7), hertwig04, '-o', color='gray', label='Hertwig (2004)', alpha=.4)
ax.plot(range(1,7), [np.mean(choices(1, prob)) for prob in range(6)], '-o', label='Exp. 1')
ax.plot(range(1,7), [np.mean(choices(2, prob)) for prob in range(6)], '-o', label='Exp. 2')
ax.set_xlim(0, 7)
ax.set_ylim(0, 1)
ax.legend()
ax.set_xlabel('Problem')
ax.set_ylabel('Proportion choosing H')
ax.set_title('Proportion choosing H across studies')
plt.show()
```

# Study 1

```python
fitdata = []
for gid in range(6):
    tofit = np.transpose((np.array(choices(1, gid)), np.array(samplesize(1, gid)), np.ones(len(choices(1, gid)))))
    
    ss_A = tofit[np.where(tofit[:,0]==0)[0],1]
    ss_B = tofit[np.where(tofit[:,0]==1)[0],1]
    max_t = np.max([np.max(ss_A), np.max(ss_B)]) + 1
    
    fitdata.append({'gid': gid,
                    'options': get_options(1, gid),
                    'max_t': max_t,
                    'samplesize': tofit})
```

### Fitting probability distortion parameters

```python
pars = {'data': fitdata,
        'z_temp': 1.,
        'fitting': ['theta', 'prelec_gamma', 'prelec_elevation']}
init = [5., 1., 1.]
f = minimize(loglik_across_gambles, init, (pars,), method='Nelder-Mead')
```

```python
print f
print 'bic: ', bic(f, pars)
```

```python
plot_probability_weighting_fcn_prelec(f['x'][2], f['x'][1])
```

```python
p_resp = []
p_stop = []
p_stop_all = []
prop_choices_predicted = []
for i, fd in enumerate(fitdata):
    pred_pars = {'data': fd['samplesize'],
                 'max_T': fd['max_t'],
                 'options': fd['options']}
    for par_i, par_name in enumerate(pars['fitting']):
        pred_pars[par_name] = f['x'][par_i]
    result = run(pred_pars)
    p_resp.append(result['resp_prob'][1])
    p_stop_all.append(result['p_stop_t'])

    exp_tA = np.sum([t * pt for (t, pt) in enumerate(result['p_stop_t'][:,0])]) * result['resp_prob'][0]
    exp_tB = np.sum([t * pt for (t, pt) in enumerate(result['p_stop_t'][:,1])]) * result['resp_prob'][1]
    p_stop.append(exp_tA + exp_tB)
    
    matched = [choice==np.argmax(result['resp_prob_t'][t]) for (choice, t, grp) in pred_pars['data']]
    prop_choices_predicted.append(np.sum(1*matched)/float(len(matched)))
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(range(1,7), [np.mean(choices(1, pid)) for pid in range(6)], '-o', label='observed')
ax.plot(range(1,7), p_resp, '-x', label='best fit model')
ax.set_xlim(0, 7)
ax.set_ylim(0, 1)
ax.set_xlabel('Problem')
ax.set_ylabel('Prob. choosing H')
ax.legend()
plt.show()
print np.mean(prop_choices_predicted)
```

```python
fig, ax = plt.subplots(figsize=(10,4))
ax.plot(range(1,7), map(np.mean, ss1), '-o', label='observed')
ax.plot(range(1,7), p_stop, '-x', label='best fit model')
ax.legend()
ax.set_xlim(0, 7)
ax.set_ylim(0, 25)
ax.set_xlabel('Problem')
ax.set_ylabel('Sample size')
ax.set_title('Observed mean sample size')
plt.show()
```

```python
p_stop_combined_e1 = [np.sum(both, axis=1) for both in p_stop_all]
fig, ax = plt.subplots(2, 3, figsize=(20,6))
for gid in range(6):
    axi = ax[gid/3][gid%3]
    axi.hist(ss1[gid], bins=range(0, 180, 5), alpha=.5, label='Exp 1')
    axModel = axi.twinx()
    axModel.plot(p_stop_combined_e1[gid])

    axi.set_ylim(0, 15)
    axi.set_xlim(0, 180)
    axi.set_title('Problem %s' % (gid+1))
    axi.set_xlabel('Sample size')
plt.show()
```
