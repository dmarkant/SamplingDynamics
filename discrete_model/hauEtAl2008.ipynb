# Hau et al (2008)

In this paper there were three experiments that used the original set of problems from Hertwig et al (2004). 

- Exp 1: Direct replication of Hertwig et al. (2004)
- Exp 2: Increased stakes 
- Exp 3: Increased sample size (fixed at 100 per problem)

Here going to fit the model to the results of Experiment 1 and 2. First will fit models separately, then will test for a difference between Experiment 1 and 2 by combining those datasets.

### Switching rate and sample size

```python
sdata = hau2008.sampledata_by_subject(study=1)
SUBJ = np.sort(sdata.keys())
```

```python
def switch_counts(study, sid):
    sr = []
    for gid in range(6):
        samp = hau2008.get_subj_data(study, sid, gid)['samples']
        if len(samp)==0:
            sr.append(np.nan)
        else:
            switches = 0
            for i in range(1, len(samp)):
                if samp[i]!=samp[i-1]:
                    switches += 1
            sr.append(switches)
    return sr
def switch_rates(study, sid):
    sr = []
    for gid in range(6):
        samp = hau2008.get_subj_data(study, sid, gid)['samples']
        if len(samp)==0:
            sr.append(np.nan)
        else:
            switches = 0
            for i in range(1, len(samp)):
                if samp[i]!=samp[i-1]:
                    switches += 1
            sr.append(switches / float(len(samp)-1))
    return sr
def sample_sizes(study, sid):
    ss = []
    for gid in range(6):
        samp = hau2008.get_subj_data(study, sid, gid)['samples']
        if len(samp)==0:
            ss.append(np.nan)
        else:
            ss.append(len(samp))
    return ss
```

```python
avg_sf = {study: [np.nanmean(switch_rates(study, sid)) for sid in hau2008.subjects(study)] for study in [1, 2]}
avg_ss = {study: [np.nanmean(sample_sizes(study, sid)) for sid in hau2008.subjects(study)] for study in [1, 2]}
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16,5))
for study in [1, 2]:
    ax[study-1].hist(avg_sf[study], bins=30, alpha=0.5, color='gray')
    ax[study-1].set_title('Average switching frequency\nStudy %s' % study)
    ax[study-1].set_xlabel('Average switching frequency')
    ax[study-1].set_ylabel('# participants')
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16,5))
for study in [1, 2]:
    ax[study-1].plot(avg_ss[study], avg_sf[study], 'kx', markersize=10)
    ax[study-1].set_title('Switching frequency by sample size\nStudy %s' % study)
    ax[study-1].set_xlabel('Average sample size')
    ax[study-1].set_ylabel('Average switching frequency')
plt.show()
```

Okay but does sample size or switching frequency relate to the proportion of H choices?

```python
fitline = {}
sampledata = {}
for sid in hau2008.subjects(study=1):
    sc = filter(lambda sc: ~np.isnan(sc), switch_counts(1, sid))
    ss = filter(lambda ss: ~np.isnan(ss), sample_sizes(1, sid))
    sampledata[(1, sid)] = {'switchcount': sc, 'samplesize': ss}
    fitline[(1, sid)] = np.polyfit(ss, sc, 1)
```
