```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mypy.explib.frey2014 import frey2014
from model_natmean import model_natmean
%load_ext autoreload
%autoreload 2
```

```python
subjdata = frey2014.sampledata_by_subject()
```

```python
sid = 36
```

```python
ret = model_natmean.fit_subject_across_gambles(subjdata[sid], 
                                               {}, 
                                               ['theta', 'eval_pow', 'eval_crit', 'z_sd', 'rho'])
```

```python
ret
```

```python
states = []
for sdata in subjdata[sid]:
    pars = {'data': sdata}
    for p in ret['bf_par']:
        pars[p] = ret['bf_par'][p]
    result = model_natmean.run(pars)
    states.append(result['pref'])
```

```python
subjdata[sid][1]
```

```python
fig, ax = plt.subplots(figsize=(20,5))
for i, gstate in enumerate(states):
    ax.plot(gstate, color='blue' if subjdata[sid][i]['choice']==0 else 'red', alpha=.2)
ax.plot([0, 30], [ret['bf_par']['theta'], ret['bf_par']['theta']], color='black')
ax.plot([0, 30], [-ret['bf_par']['theta'], -ret['bf_par']['theta']], color='black')
```
