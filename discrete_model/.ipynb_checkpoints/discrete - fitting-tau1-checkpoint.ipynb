```python
# create an artificial dataset, then try to fit it
pars = {'theta': 5.,
        'delta': -0.05,
        'z_temp': 0.5,
        'max_T': 100,
        'verbose': True}
predicted = run(pars)
```

```python
T = predicted['T']
p_A = predicted['resp_prob'][0]
print p_A
N = 5000
prop_A = np.random.binomial(N, p_A)
prop_B = N - prop_A
```

```python
resp_prob_t = predicted['resp_prob_t']
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(T, resp_prob_t[:,0], label='A')
ax.plot(T, resp_prob_t[:,1], label='B')
ax.set_title('Choice probability over time')
plt.show()
```

```python
pred = predicted['p_stop_t'].transpose()
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(pred[0,:])
ax.plot(pred[1,:])
plt.show()
```

```python
# for A choices, sample corresponding sample size
cs_A = predicted['p_stop_t_cump'][:,0]
ss_A = [np.sum(np.random.random() > cs_A)-1 for _ in range(prop_A)]

cs_B = predicted['p_stop_t_cump'][:,1]
ss_B = [np.sum(np.random.random() > cs_B)-1 for _ in range(prop_B)]

data = np.concatenate(([[0, s] for s in ss_A],
                       [[1, s] for s in ss_B]))
```

```python
fig, ax = plt.subplots(1, 2)
ax[0].plot(T, cs_A)
ax[1].hist(ss_A, bins=30)
plt.show()
```

## All problems, single drift

```python
grid_fit_pars = {}
grid_fit_results = {}
grid_bf_theta = {}
THETAS = np.arange(1, 10)
```

```python
for gid in gamble_lab[50:]:
    print gid
    gdata = data[data['gamble_lab']==gid]
    tofit = []
    for pair in gdata[['decision', 'samplesize']].values:
        choice = 0 if pair[0]=='L' else 1
        samplesize = pair[1]
        tofit.append([choice, samplesize])
    tofit = np.array(tofit)
    
    ss_A = tofit[np.where(tofit[:,0]==0)[0],1]
    ss_B = tofit[np.where(tofit[:,0]==1)[0],1]
    max_t = np.max([np.max(ss_A), np.max(ss_B)]) + 1
    
    for theta in THETAS:
        
        pars = {'data': tofit,
                'theta': theta,
                'max_T': max_t,
                'fitting': ['delta', 'z_temp'],
                'verbose': False}
        
        init = [np.random.random()-1, np.random.random()*2]
        grid_fit_pars[gid] = pars
        grid_fit_results[(gid, theta)] = minimize(loglik, init, (pars,), method='Nelder-Mead')
        #print gid
        #print 'init:', init
        #print fit_results[(gid, theta)]

    # get results for best theta
    grid_bf_theta[gid] = THETAS[np.argmin([grid_fit_results[(gid, th)]['fun'] for th in THETAS])]
```

```python
lh_by_theta = np.array([[grid_fit_results[(gid, th)]['fun'] for th in THETAS] for gid in gamble_lab])
```

```python
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(THETAS, lh_by_theta.transpose(), color='gray', alpha=.3)
plt.show()
```

```python
for gid in gamble_lab:
    pars = grid_fit_pars[gid]
    
    fr = grid_fit_results[(gid, grid_bf_theta[gid])]
    pars['theta'] = grid_bf_theta[gid]
    for i, p in enumerate(pars['fitting']):
        pars.update({p: fr['x'][i]})

    print gid, pars['theta'], pars['delta'], pars['z_temp']
    
    bf_pred = run(pars)
    
    tofit = pars['data']
    ss_A = tofit[np.where(tofit[:,0]==0)[0],1]
    ss_B = tofit[np.where(tofit[:,0]==1)[0],1]
    
    fig, ax = plt.subplots(1, 2, figsize=(10,3))
    ax[0].plot(bf_pred['T'], bf_pred['resp_prob_t'][:,0], label='A')
    ax[0].set_title('%s: Chose A' % gid)
    axhA = ax[0].twinx()
    axhA.hist(ss_A, bins=bf_pred['T'], alpha=.9)
    ax[0].set_xlim([0, 200])

    ax[1].plot(bf_pred['T'], bf_pred['resp_prob_t'][:,1], label='B')
    ax[1].set_title('%s: Chose B' % gid)
    axhB = ax[1].twinx()
    axhB.hist(ss_B, bins=bf_pred['T'], alpha=.9)
    ax[1].set_xlim([0, 200])
    plt.show()   
```

## All gambles, two groups

```python
grid_fit_pars_2 = {}
grid_fit_results_2 = {}
grid_bf_theta_2 = {}
THETAS = np.arange(1, 10)
```

```python
for gid in gamble_lab:
    print gid
    gdata = data[data['gamble_lab']==gid]
    
    tofit = {'old': [], 'yng': []}
    ss_A = {}
    ss_B = {}
    max_t = {}
    for grp in ['old', 'yng']:
        gdata_grp = gdata[gdata['group']==grp]
        for pair in gdata_grp[['decision', 'samplesize']].values:
            choice = 0 if pair[0]=='L' else 1
            samplesize = pair[1]
            tofit[grp].append([choice, samplesize])
        tofit[grp] = np.array(tofit[grp])

        ss_A[grp] = tofit[grp][np.where(tofit[grp][:,0]==0)[0],1]
        ss_B[grp] = tofit[grp][np.where(tofit[grp][:,0]==1)[0],1]
        max_t[grp] = np.max([np.max(ss_A[grp]), np.max(ss_B[grp])]) + 1


        for theta in THETAS:

            pars = {'data': tofit[grp],
                    'theta': theta,
                    'z_temp': 0.5,
                    'max_T': max_t[grp],
                    'fitting': ['delta'],
                    'verbose': False}

            init = [np.random.random()-1]
            grid_fit_pars_2[(gid, grp)] = pars
            grid_fit_results_2[(gid, grp, theta)] = minimize(loglik, init, (pars,), method='Nelder-Mead')
            #print gid
            #print 'init:', init
            #print grid_fit_results_2[(gid, grp, theta)]

        # get results for best theta
        grid_bf_theta_2[(gid, grp)] = THETAS[np.argmin([grid_fit_results_2[(gid, grp, th)]['fun'] for th in THETAS])]
```

```python
for gid in gamble_lab[:10]:
    fig, ax = plt.subplots(1, 4, figsize=(10,2))

    for gi, grp in enumerate(['old', 'yng']):
        pars = grid_fit_pars_2[(gid, grp)]
    
        fr = grid_fit_results_2[(gid, grp, grid_bf_theta_2[(gid, grp)])]
        pars['theta'] = grid_bf_theta_2[(gid, grp)]
        for i, p in enumerate(pars['fitting']):
            pars.update({p: fr['x'][i]})

        print grp, gid, pars['theta'], pars['delta'], pars['z_temp']

        bf_pred = run(pars)

        tofit = pars['data']
        ss_A = tofit[np.where(tofit[:,0]==0)[0],1]
        ss_B = tofit[np.where(tofit[:,0]==1)[0],1]
        
        ind = gi*2
        ax[ind].plot(bf_pred['T'], bf_pred['resp_prob_t'][:,0], label='A')
        ax[ind].set_title('%s, %s:\nChose A' % (grp, gid))
        axhA = ax[ind].twinx()
        axhA.hist(ss_A, bins=bf_pred['T'], alpha=.9)
        ax[ind].set_xlim([0, 200])

        ax[ind+1].plot(bf_pred['T'], bf_pred['resp_prob_t'][:,1], label='B')
        ax[ind+1].set_title('%s, %s:\nChose B' % (grp, gid))
        axhB = ax[ind+1].twinx()
        axhB.hist(ss_B, bins=bf_pred['T'], alpha=.9)
        ax[ind+1].set_xlim([0, 200])
    plt.show() 
```

```python
gset = gamble_lab_srt
fig, ax = plt.subplots(2, 1, figsize=(20,6))
for i, grp in enumerate(['old', 'yng']):
    col = 'red' if grp=='old' else 'blue'
    bf_theta = [grid_bf_theta_2[(gid, grp)] for gid in gset]
    ax[0].bar(range(len(bf_theta)), bf_theta, color=col, alpha=.5)
    ax[0].set_ylim([0, 8])
    ax[0].set_xlim([0, 85])
    ax[0] = add_problem_labels(ax[0], 7, [0, 8])

    bf_delta = [grid_fit_results_2[(gid, grp, grid_bf_theta_2[(gid, grp)])]['x'][0] for gid in gset]
    ax[1].bar(range(len(bf_delta)), bf_delta, color=col, alpha=.5)
    ax[1].set_ylim([-.35, .25])
    ax[1].set_xlim([0, 85])
    ax[1] = add_problem_labels(ax[1], -.3, [-.35, .25])

plt.show()
```
