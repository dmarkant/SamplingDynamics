```python
cols = ['blue', 'red']
fig, ax = plt.subplots(figsize=(20,8))
for gind, gid in enumerate(range(6)):
    for opt_ind, opt in enumerate(get_options(gid)):
        for x, p in opt:
            ax.plot(gind+1, x, '.', color=cols[opt_ind], markersize=(85*p), alpha=.7)
ax.set_xlim([0, 7])
ax.set_xlabel('Problem')
ax.set_ylabel('Outcome')
ax.set_title('Outcomes for all problems (Red=H, Blue=L), with size of marker indicating probability')
plt.show()
```

```python
ss = [samplesize(pid) for pid in range(6)]
```

```python
fig, ax = plt.subplots(1, 6, figsize=(20,4))
for gid in range(6):
    ax[gid].hist(ss[gid], bins=range(0, 100, 10), alpha=.5)
    ax[gid].set_ylim(0, 25)
    ax[gid].set_xlim(0, 100)
    ax[gid].set_title('Problem %s' % (gid+1))
    ax[gid].set_xlabel('Sample size')
plt.show()
```

```python
fitdata = []
for gid in range(6):
    tofit = np.transpose((np.array(choices(gid)), np.array(samplesize(gid)), np.ones(len(choices(gid)))))
    
    ss_A = tofit[np.where(tofit[:,0]==0)[0],1]
    ss_B = tofit[np.where(tofit[:,0]==1)[0],1]
    max_t = np.max([np.max(ss_A), np.max(ss_B)]) + 1
    
    fitdata.append({'gid': gid,
                    'options': get_options(gid),
                    'max_t': max_t,
                    'samplesize': tofit})
```

```python
pars = {'data': fitdata,
        'fitting': ['theta', 'z_temp', 'prelec_gamma', 'prelec_elevation'],
        'verbose': True
}
init = map(randstart, pars['fitting'])
loglik_across_gambles(init, pars)
```

```python
f = minimize(loglik_across_gambles, init, (pars,), method='Nelder-Mead')
```

```python
print f
print 'bic: ', bic(f, pars)
```

```python
p_resp = []
p_stop = []
prop_choices_predicted = []
for i, fd in enumerate(fitdata):
    pred_pars = {'data': fd['samplesize'],
                 'max_T': fd['max_t'],
                 'options': fd['options']}
    for par_i, par_name in enumerate(pars['fitting']):
        pred_pars[par_name] = f['x'][par_i]
    result = run(pred_pars)
    p_resp.append(result['resp_prob'][1])
    
    exp_tA = np.sum([t * pt for (t, pt) in enumerate(result['p_stop_t'][:,0])]) * result['resp_prob'][0]
    exp_tB = np.sum([t * pt for (t, pt) in enumerate(result['p_stop_t'][:,1])]) * result['resp_prob'][1]
    p_stop.append(exp_tA + exp_tB)
    
    matched = [choice==np.argmax(result['resp_prob_t'][t]) for (choice, t, grp) in pred_pars['data']]
    prop_choices_predicted.append(np.sum(1*matched)/float(len(matched)))
```
