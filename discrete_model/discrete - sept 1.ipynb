# The basic model

### The state space

The state space is defined as discrete levels of evidence between two decision thresholds, $\theta$ and $-\theta$. Given a function that specifies the probability of transitioning up or down (more on that below), Diederich (1997) shows how to compute the choice probabilities $p(choice=A)$ and $p(choice=B)$, as well as the probability of stopping at a particular time conditioned on which alternative is chosen, $p(t|choice)$.

For example, for $\theta=5$ we have a state space ranging from -5 to 5. The plot on the left below shows the probability of occupying these states over time. Note that the initial state vector $Z$ is peaked at the unbiased middle point of the state space. The middle curve shows the probability of making each choice given at each timestep, whereas the right curve shows the probability of stopping at each timestep given the chosen option (in this case, they distribution of stopping times is the same).

```python
result = run(pars = {'theta': 5.,
                     'delta': -0.1,
                     'z_temp': 0.5})
plot_model(result)
```

Increasing the threshold $\theta$ means that more evidence is required to make a decision. The state space increases in size (i.e., more steps have to be traversed), and the curves are pushed foward in time.

```python
result = run(pars = {'theta': 10,
                     'delta': -0.1,
                     'z_temp': 0.5})
plot_model(result)
```

The vector $Z$ specifies the the probability distribution over starting states, which is generally assumed to be unbiased. For example, could assume that the process starts at the midpoint with probability of 1. Alternatively, could assume a uniform distribution over starting positions. In my version I wanted something that could vary between those two extremes, so the probability of a starting position is based on its distance from the midpoint, then passed through a softmax function controlled by a temperature parameter $z_{temp}$. When $z_{temp}$ is low (as above) the starting position is highly likely to be in the center. As $z_{temp}$ increases, starting positions further away from the center are increasingly likely.

For example, if $z_{temp}$ is high there will be a higher proportion of fast decisions (since people are more likely to start close to a threshold). This can allow for fast errors. As seen in the plot below on the right, if an incorrect choice is made,  it is most likely to occur early on due to the wide disperson of starting states.

```python
result = run(pars = {'theta': 10.,
                     'delta': -0.1,
                     'z_temp': 10})
plot_model(result)
```

Note that in the results above, both the step size between states and between timepoints is assumed to be 1. However, a main point of Diederich (1997) is that this discrete model can be used to approximate a time-continuous process. Decreasing the stepsize $dv$ will increase the size of the state space and the quality of the approximation to a continuous process, but doesn't have an effect on the relative response and sample size probabilities (assuming we evaluate them at the same points in time). 

For example, in the plots below $dv=.25$, such that a unit of time corresponds to 4 steps in the Markov chain. Note that the relative probabilities shown in the plots to the right are the same as in the plots above. 

For discrete sampling paradigms where the measure is the sample size (rather than continuous time), the model can thus be used assuming that each sample maps on to a step in the model. One nice property is that you can then incorporate finer grained process information (e.g., time spent looking at an outcome) within the same model and test for a better fit. (However, for now I will restrict it to the unit step case).

```python
result = run(pars = {'theta': 10.,
                     'delta': -0.1,
                     'z_temp': 10,
                     'dv': .25})
plot_model(result)
```

### Transition probabilities

The model is formulated as a birth-death chain: with each step of the chain, the state can only 1) go down one step, 2) stay in the same state, or 3) go up one step. The transition probabilities specify the likelihood of each of these actions.

$p_{down} = \frac{1}{2 \, \alpha} (1 - \frac{\delta - \gamma \cdot v}{\sigma}) \cdot \sqrt{\tau}$

$p_{up} = \frac{1}{2 \, \alpha} (1 - \frac{\delta + \gamma \cdot v}{\sigma}) \cdot \sqrt{\tau}$

$p_{stay} = 1 - \frac{1}{\alpha}$


The drift coefficient ($\delta - \gamma \cdot v$) controls the direction and velocity, with $\delta$ indicating the difference between alternatives (i.e., the result of comparing the alternatives), $x$ is the current state of the process, and $\gamma$ is a free parameter that weights the current state.

For now, the parameter we care most about is $\delta$ (and set $\gamma=0$). An increase in $\delta$ leads to faster decisions:

```python
result = run(pars = {'theta': 10.,
                     'delta': -.5,
                     'z_temp': 1,
                     'dv': .25})
plot_model(result)
```

## Fitting a single problem

Using data from Frey et al. study comparing older and younger adults (see more details below).

```python
data = pd.read_csv('../dfe_by_game.csv')
gamble_lab = data['gamble_lab'].unique()

# sort by problem type
gambles = data[data['partid']==1]
gambles_srt = gambles.sort(['domain', 'pairtype', 'session'])
gamble_lab_srt = gambles_srt['gamble_lab']

gdata = data[data['gamble_lab']=='S11G1']
```

```python
tofit = {'old': [], 'yng': []}
ss_A = {}
ss_B = {}
max_t = {}
for grp in ['old', 'yng']:
    gdata_grp = gdata[gdata['group']==grp]
    for pair in gdata_grp[['decision', 'samplesize']].values:
        choice = 0 if pair[0]=='L' else 1
        samplesize = pair[1]
        tofit[grp].append([choice, samplesize])
    tofit[grp] = np.array(tofit[grp])

    ss_A[grp] = tofit[grp][np.where(tofit[grp][:,0]==0)[0],1]
    ss_B[grp] = tofit[grp][np.where(tofit[grp][:,0]==1)[0],1]
    max_t[grp] = np.max([np.max(ss_A[grp]), np.max(ss_B[grp])]) + 1

fig, ax = plt.subplots(1, 2, figsize=(20,4))
for i, grp in enumerate(['old', 'yng']):
    ax[i].hist(ss_A[grp], range(max_t[grp]), label='chose L')
    ax[i].hist(ss_B[grp], range(max_t[grp]), label='chose H')
    ax[i].set_title(grp)
    ax[i].set_xlabel('Sample size')
    ax[i].set_ylabel('# of participants')
    ax[i].legend()
plt.show()
```

```python
# fitting theta, delta separately for each group
f = {}
for grp in ['old', 'yng']:
    pars = {'data': tofit[grp],
            'z_temp': 1,
            'max_T': max_t[grp],
            'fitting': ['theta', 'delta']}

    init = [np.random.randint(2, 10), 0.]
    f[grp] = minimize(loglik, init, (pars,), method='Nelder-Mead')

# given best-fit parameters, get predicted sample sizes
bf_pred = {}
for grp in ['old', 'yng']:
    bf_pred[grp] = run({'theta': f[grp]['x'][0],
                        'delta': f[grp]['x'][1],
                        'z_temp': 0.5,
                        'max_T': max_t[grp]})
```

```python
tmax = np.max([max_t[grp] for grp in ['old', 'yng']])
fig, ax = plt.subplots(2, 2, figsize=(20,5))
for i, grp in enumerate(['old', 'yng']):
    ax[i][0].plot(bf_pred[grp]['T'], bf_pred[grp]['resp_prob_t'][:,0], label='A')
    ax[i][0].set_title('%s, chose L' % grp)
    axhA = ax[i][0].twinx()
    axhA.hist(ss_A[grp], bins=range(max_t[grp]), alpha=.7)
    ax[i][0].set_xlim([0, tmax])
    
    ax[i][1].plot(bf_pred[grp]['T'], bf_pred[grp]['resp_prob_t'][:,1], label='B')
    ax[i][1].set_title('%s, chose H' % grp)
    axhB = ax[i][1].twinx()
    axhB.hist(ss_B[grp], bins=range(max_t[grp]), alpha=.7)
    ax[i][1].set_xlim([0, tmax])
    
plt.show()
```

## Fitting all 84 problems

Now want to look at all of the problems in the aging dataset. Summary of data:

- Two groups of participants: (1) 35 younger adults, (2) 35 older adults
- Each participant completed 3 sessions a day for 7 consecutive days
- Each session included 4 decision problems, 2 in gain domain and 2 in loss domain
- At end of each session, received feedback about total gains/losses

Four sets of options:

1. low expected value, low variance (LL)
2. low expected value, high variance (LH)
3. high expected value, low variance (HL)
4. high expected value, high variance (HH)

Decision problems involved pairing

1. HH vs. LL 
2. HL vs. LH

and were either in the gain or loss domain.

```python
data_by_gamble = pd.read_csv('../dfe_by_gamble.csv')

def get_options(gid):
    gdata = data_by_gamble[data_by_gamble['gid']==gid]
    L = np.array([gdata[['L_x1', 'L_p1']].values, gdata[['L_x2', 'L_p2']].values, gdata[['L_x3', 'L_p3']].values]).reshape((3,2))
    H = np.array([gdata[['H_x1', 'H_p1']].values, gdata[['H_x2', 'H_p2']].values, gdata[['H_x3', 'H_p3']].values]).reshape((3,2))
    return [L, H]
```

```python
cols = ['blue', 'red']
fig, ax = plt.subplots(figsize=(20,8))
for gind, gid in enumerate(gamble_lab_srt):
    for opt_ind, opt in enumerate(get_options(gid)):
        for x, p in opt:
            ax.plot(gind, x, '.', color=cols[opt_ind], markersize=(35*p), alpha=.4)
ax.set_xlim([-1, 85])
add_problem_labels(ax, -28, [-30, 30])
ax.set_xlabel('Problem')
ax.set_ylabel('Outcome')
ax.set_title('Outcomes for all problems (Red=H, Blue=L), with size of marker indicating probability')
plt.show()
```

There are a couple of interesting patterns in the results that should be reflected in the model fits. First, there was an interaction between domain and pair type on choice accuracy (proportion of people choosing the higher option, H). 

For gains, accuracy was higher for HL-LH problems (in which the H option had lower variance). For losses the pattern was reversed, with accuracy higher for HH-LL problems (in which the H option had high variance). This seems to have occurred because the gain gambles were reflected across zero to create the loss gambles, and as a result the rare, extreme outcomes (really, the lack thereof due to small samples) have different consequences on choices.

```python
fig, ax = plt.subplots(figsize=(20,5))
ax.plot(data_by_gamble['prop_H_all_old'].values, 'r-x', label='old')
ax.plot(data_by_gamble['prop_H_all_yng'].values, 'b-x', label='young')
add_problem_labels(ax, .1, [0,1])
ax.legend()
ax.set_xlim(-1, 85)
ax.set_xlabel('Problem')
ax.set_ylabel('Proportion choosing H')
ax.set_title('Observed accuracy (proportion of people choosing H option)')
plt.show()
```

Second, there was a qualitative difference in sample sizes between groups, with older adults drawing fewer samples across all problem types. Note that this difference was not significant based on either Renato's or my mixed effects modeling, but as can be seen below it is very consistent across problems.

```python
fig, ax = plt.subplots(figsize=(20,5))
ax.plot(data_by_gamble['ss_all_old'].values, 'r-+', label='old')
ax.plot(data_by_gamble['ss_all_yng'].values, 'b-+', label='young')
add_problem_labels(ax, 14, [0,50])
ax.legend()
ax.set_xlim(-1, 85)
ax.set_ylim(10, 45)
ax.set_xlabel('Problem')
ax.set_ylabel('Sample size')
ax.set_title('Observed mean sample size')
plt.show()
```

### Modeling drift rate with CPT weighting functions

Next question is whether we can account for the results based on the structure of the gambles, rather than fitting parameters spearately for each problem. My first pass at this was to model the drift rate using the difference in expected utility a la CPT. Here there are a few additional parameters to consider:

$\delta$: scaling factor (before this was the drift rate itself)

$w_p$: probability weighting parameter

$pow_{gain}$, $pow_{loss}$: value weighting parameters

$w_{loss}$: loss-specific scaling parameter

```python
from model_cpt import *
gset = gamble_lab_srt
```

```python
fitdata = []
for gid in gset:
    gdata = data[data['gamble_lab']==gid]

    tofit = []
    for obs in gdata[['decision', 'samplesize', 'group']].values:
        choice = 0 if obs[0]=='L' else 1
        group = 0 if obs[2]=='old' else 1
        samplesize = obs[1]
        tofit.append([choice, samplesize, group])
    tofit = np.array(tofit)
    
    ss_A = tofit[np.where(tofit[:,0]==0)[0],1]
    ss_B = tofit[np.where(tofit[:,0]==1)[0],1]
    max_t = np.max([np.max(ss_A), np.max(ss_B)]) + 1
    
    fitdata.append({'gid': gid,
                    'options': get_options(gid),
                    'max_t': max_t,
                    'samplesize': tofit})
```
